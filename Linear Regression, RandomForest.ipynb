## **Load data**

```python
import pandas as pd
data= pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
```

```python
data= pd.read_csv('delaney_solubility_with_descriptors.csv')
```

```python
data
```

### **Data Preparation**

#### **Data Separation as X and y**

```python
y= data['logS']
y
```

```python
X=data.drop('logS', axis=1)
X
```

#### **Data splitting**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2, random_state=100)
```

```python
X_train, X_test, y_train, y_test
```

## **Model Build**

### **Linear Regression**

#### **Training the model**

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_train,y_train)
```

#### **Applying the model to make prediction**

```python
y_lr_train_pred=lr.predict(X_train)
y_lr_test_pred=lr.predict(X_test)
```

```python
y_lr_train_pred, y_lr_test_pred
```

#### **Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr_train_rmse=mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2=r2_score(y_train, y_lr_train_pred) 
```

```python
lr_test_rmse=mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2=r2_score(y_test, y_lr_test_pred) 
```

```python
print('LR RMSE (Train):', lr_train_rmse)
print('LR R2 (Train):' ,lr_train_r2)
print()
print()
print('LR RMSE (Test):', lr_test_rmse)
print('LR R2 (Test):' ,lr_test_r2)
```

```python
import pandas as pd
lr_results= pd.DataFrame(['LinearRegression',  lr_train_rmse, lr_train_r2,lr_test_rmse,lr_test_r2 ]). transpose()
lr_results.columns= ['Method', 'Train RMSE', 'Train RÂ²', 'Test RMSE', 'Test RÂ²']
lr_results
```

### **Random Forest** 


#### **Training The model**

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

#### **Applying the model to make prediction**

```python
y_rf_train_pred=rf.predict(X_train)
y_rf_test_pred=rf.predict(X_test)
```

#### **Evaluate Model Performance**

```python
rf_train_rmse=mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2=r2_score(y_train, y_rf_train_pred) 
```

```python
rf_test_rmse=mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2=r2_score(y_test, y_rf_test_pred) 
```

```python
print('RF RMSE (Train):', rf_train_rmse)
print('RF R2 (Train):' ,rf_train_r2)
print()
print()
print('RF RMSE (Test):', rf_test_rmse)
print('RF R2 (Test):' ,rf_test_r2)
```

```python
import pandas as pd
rf_results= pd.DataFrame(['RandomForestRegression',  rf_train_rmse, rf_train_r2,rf_test_rmse,rf_test_r2 ]). transpose()
rf_results.columns= ['Method', 'Train RMSE', 'Train RÂ²', 'Test RMSE', 'Test RÂ²']
rf_results
```

### **Model Comparison**

```python
data_models=pd.concat((lr_results, rf_results), axis=0)
data_models
```

```python
data_models.reset_index(drop=True)
```

### **Data visualization of prediction results**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.scatter(x=y_train, y=y_lr_train_pred, c='green', alpha=0.5)

z=np.polyfit(y_train, y_lr_train_pred, 1)
p=np.poly1d(z)

plt.plot(y_train, p(y_train), c='red')

plt.ylabel('Predict logs')
plt.xlabel('Experimental logs')
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Choose one tree from the forest
estimator = rf.estimators_[0]  
# Plot the tree
plt.figure(figsize=(20, 10))  
plot_tree(estimator,
          feature_names=X_train.columns,
          filled=True,
          rounded=True,
          fontsize=10)
plt.title("Visualization of One Tree from the Random Forest")
plt.show()
```
